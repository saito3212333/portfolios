```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/Users/sahiyo/PycharmProjects/Data/athlete_events.csv')
```

```python
df = df.dropna(subset=['Medal'])
```

```python
top_sport = ["Athletics",'Swimming','Rowing',"Football",'Hockey']
```

```python
df1 = df[(df.Sport.isin(top_sport)) & (df.Year == 2016)] 
```

```python
df1.head()
```

```python
df1.info()
```

```python
sns.scatterplot(x="Height", y="Weight",data=df1)
```

```python
sns.jointplot(x="Height", y="Weight",data=df1,kind="hex")
```

```python
ax1 = sns.boxplot(x="Medal", y="Weight",data=df1,hue="Sex")
```

```python
ax2 = sns.violinplot(x="Medal", y="Weight",data=df1,hue="Sex")
```

```python
ax3 = sns.violinplot(x="Sport", y="Weight",data=df1,hue="Sex")
```
