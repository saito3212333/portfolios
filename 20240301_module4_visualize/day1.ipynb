```python
import pandas as pd
import seaborn as sbn
```

```python
mpg_df = sbn.load_dataset('mpg')
mpg_df.head()
```

```python
ax = sbn.scatterplot(x='weight',y='mpg',data=mpg_df)
```

```python
ax = sbn.jointplot(x='weight',y='mpg',data=mpg_df,kind='hex',color='blue')
```

```python
ax = sbn.kdeplot(x='weight',y='mpg',data=mpg_df)
```

```python
ax = sbn.scatterplot(x='model_year',y='acceleration',data=mpg_df)
```

```python
ax = sbn.lineplot(x='model_year',y='acceleration',data=mpg_df)
```

```python
flights_df = sbn.load_dataset('flights')
flights_df.tail(20)
```

```python
for i in flights_df['month'].unique():
    ax = sbn.lineplot(x='year',y='passengers',data=flights_df[flights_df['month']==i])
```

```python
ax = sbn.lineplot(x='year',y='passengers',data=flights_df,hue='month')
```

```python
ax = sbn.lineplot(x='month',y='passengers',data=flights_df,hue='year')
```

```python
df_pivoted = flights_df.pivot(columns='month',index='year',values='passengers')
ax = sbn.heatmap(df_pivoted)
```

```python
import matplotlib.pyplot as plt
import numpy as np
ax = sbn.histplot(mpg_df.weight,bins=50)
plt.axvline(x=np.mean(mpg_df.weight),color='red',label='mean')
plt.axvline(x=np.median(mpg_df.weight),color='yellow',label='median')
plt.legend(loc='upper right')
```
