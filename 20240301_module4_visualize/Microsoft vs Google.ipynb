```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from  bokeh.plotting import figure, output_notebook, show, ColumnDataSource
from bokeh.io import output_notebook, show,push_notebook
```

```python
output_notebook()
```

```python
def prepare_data():
    df_mi = pd.read_csv('/Users/sahiyo/PycharmProjects/Data/microsoft_stock.csv')
    df_mi['date'] = pd.to_datetime(df_mi['date'])
    df_go = pd.read_csv('/Users/sahiyo/PycharmProjects/Data/google_stock_ex6.csv')
    df_go['date'] = pd.to_datetime(df_go['date'])
    
    return df_mi,df_go
```

```python
Tooltips = [
    ('date','@date'),
    ("value",'@close')
]

p = figure(title = "Stock Comparison of Google and Microsoft",height=500,width=800, x_axis_type = 'datetime',tooltips=Tooltips)
```

```python
df_mi,df_go = prepare_data()
```

```python
m_line = p.line(x='date',y='close',source=df_mi,color='#0000FF',legend_label="Microsoft",line_width=2)
g_line = p.line(x='date',y='close',source=df_go,color='#00FF00',legend_label="Google",line_width=2)
```

```python
show(p)
```

```python
from ipywidgets import interact
def update(color,width=1):
    g_line.glyph.line_color = color
    g_line.glyph.line_width = width
    push_notebook()
interact(update,color=['red','green','blue'],line_width=1.5)
show(p,notebook_handle=True)
```
