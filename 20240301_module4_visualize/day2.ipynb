```python
import seaborn as sns
```

```python
df_athlete.describe()
```

```python
olympics_winners = df_athlete.dropna(subset=["Medal"])
```

```python
# Create a list of 
top_sports = ["Athletics", "Swiming", "Rowing", "Football","Hockey"]

olympics_winners_final = olympics_winners[(olympics_winners.Sport.isin(top_sports)) & (olympics_winners.Year == 2016)]
olympics_winners_final.head()
```

```python
olympics_winners_final.describe()
```

```python
sns.scatterplot(x="Height", y="Weight",hue="Medal", data=olympics_winners_final)
```

```python
sns.jointplot(x="Height", y="Weight"
              , data=olympics_winners_final, kind="hex")
```
